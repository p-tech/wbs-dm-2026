# STEP 5: Inspect the website
Download the page



```python
url = "http://quotes.toscrape.com"
response = requests.get(url)
response.encoding = "utf-8"

print(response.status_code)
```

Response codes:

*   200 - OK - returned, loaded
*   301, 302 - redirected page
*   404 - error, page not found
*   500 - server error



# Extract Quotes

```python
# Step 2: Parse the HTML
soup = BeautifulSoup(response.text, "html.parser")

# Step 3: Find all quote text elements
quotes = soup.find_all("span", class_="text")

# Step 4: Print quotes
for quote in quotes:
    print(quote.text)
```

# Step 5 – Extend the Task (Very Basic Challenge)

Modify the code to also extract:

The author name

The tags for each quote

Hint: Look again in Inspect and find the relevant class names.

```python
# Step 3: Find all quote blocks
quotes = soup.find_all("div", class_="quote")

# Step 4: Loop through each quote
for quote in quotes:

    # Extract quote text
    text = quote.find("span", class_="text").text

    # Extract author
    author = quote.find("small", class_="author").text

    # Extract tags (multiple possible)
    tag_elements = quote.find_all("a", class_="tag")
    tags = [tag.text for tag in tag_elements]

    # Print results
    print("Quote:", text)
    print("Author:", author)
    print("Tags:", ", ".join(tags))
    print("-" * 50)
```
