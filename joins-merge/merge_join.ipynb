Merge in python: Simple example

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [24, 30, 28]
})

# Perform an inner merge on the 'ID' column
result = pd.merge(df1, df2, on='ID', how='inner')
print("inner")
print(result)
```

```python
# Perform an inner merge on the 'ID' column
result = pd.merge(df1, df2, on='ID', how='outer')
print("Outer")
print(result)
```

```python
# Perform an inner merge on the 'ID' column
result = pd.merge(df1, df2, on='ID', how='left')
print("left")
print(result)
```

```python
# Perform an inner merge on the 'ID' column
result = pd.merge(df1, df2, on='ID', how='right')
print("inner")
print(result)
```

Join in Python

```python
df3 = pd.DataFrame({
    'Age': [24, 30, 28]
}, index=[2, 3, 4])

df4 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Perform a left join using the index
result = df4.join(df3, how='left')
print(result)
```

```python
# Perform a left join using the index
result = df4.join(df3, how='right')
print(result)
```

```python
# Perform a left join using the index
result = df4.join(df3, how='inner')
print(result)
```

```python
# Perform a left join using the index
result = df4.join(df3, how='outer')
print(result)
```
